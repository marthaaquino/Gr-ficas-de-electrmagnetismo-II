### Potencial de carga debida a una carga q entre dos plancas paralelas, con $\rho>>0$  esta dada por:
$\phi(\rho,z)={\frac{q}{\epsilon_0 d}}\exp (-\pi\rho/d)\sin(\pi z/d)  $

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
```

Obteniendo el potencial cuando d=1m, q=e

```python
d=1
e=8.8e-12
q=-1.6e-19

x = np.linspace(50,70,50)
y = np.linspace(0,d,50)

X,Y = np.meshgrid(x,y)
```

```python
def f (x,y):
    
    return (q/(d*e))*np.exp(-np.pi*x/d)*np.sin(np.pi*y/d)
```

```python
from matplotlib import cm


#Axes3D(plt.figure()).plot_surface(X,Y,f(X,Y), rstride=1, cstride=1, cmap = cm.cool)
fig = plt.figure()
ax = Axes3D(fig)
surf=ax.plot_surface(X,Y, f(X,Y), rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=True)
plt.title('Potencial [V]', fontsize=15)
plt.xlabel(r'$\rho$ [m]')  
plt.ylabel('z [m]')
fig.colorbar(surf, shrink=0.60, aspect=10, fraction=0.05)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from pylab import *
from pylab import *

#%config InlineBackend.figure_format = 'retina'

x = np.linspace(50,55,50)
y = np.linspace(0,1,50)
X, Y = np.meshgrid(x, y)


def campo_E_V_xy_1(x,y,d=1):
    #Ex,Ey = 0,0
       # Ex =-np.pi*np.exp(-np.pi*x/d)*np.sin(np.pi*y/d)/d
        #Ey =np.pi*np.exp(-np.pi*x/d)*np.cos(np.pi*y/d)/d
    return -np.pi*np.exp(-np.pi*x/d)*np.sin(np.pi*y/d)/d, np.pi*np.exp(-np.pi*x/d)*np.cos(np.pi*y/d)/d

def graf_campo_E(x,y,Ex,Ey):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)
    ax.streamplot(X, Y, Ex,Ey,density=[0.8, 0.8])
    plt.title(r'Campo $\vec{E}$')
    plt.xlabel(r'$\rho$')
    plt.ylabel('z')
    plt.xlim(50,55)
    plt.ylim(0,1)
    ax.set_aspect('equal')
   
```

```python
Ex,Ey = campo_E_V_xy_1(X,Y)

graf_campo_E(X, Y, -Ex,-Ey)
```

### Potencial elctrico debido a una carga fuera de una esfera conductora
En este caso, el potencial est dado por:
$ \phi(\vec{r})={\frac{1}{4\pi\epsilon_0}}({\frac{1}{|\vec{r}-\vec{s}|}}-{\frac{R/s}{|\vec{r}-\vec{s}R^{2}/s^{2}  |}})  $
Tomando $\vec{r}=(x,y)$ y $\vec{s}=(2R,0)$
Por lo cual el potencial se puede reescribir como:

    $\phi(x,y)= {\frac{1}{\sqrt{(x-s)^{2}+y^{2}}}} -{\frac{R}{s\sqrt{(x-R^{2}/s)^2+y^{2}}}}$

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
e=8.8e-12
q=-1.6e-19
R=1
s=2*R

x = np.linspace(R,30,50)
y = np.linspace(R,30,50)

X,Y = np.meshgrid(x,y)
def f1 (x,y):
    return (1/(4*np.pi*e))*((1/(((x-s)**2+y*y)**(1/2))-((R/s)/(((x-(s*R*R/(s*s)))**(2)+y*y)**(1/2))) ))

from matplotlib import cm


#Axes3D(plt.figure()).plot_surface(X,Y,f(X,Y), rstride=1, cstride=1, cmap = cm.cool)
fig = plt.figure()
ax = Axes3D(fig)
surf=ax.plot_surface(X,Y, f1(X,Y), rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=True)
plt.title('Potencial [V]', fontsize=15)
plt.xlabel(r'$\rho$ [m]')  
plt.ylabel('z [m]')
fig.colorbar(surf, shrink=0.60, aspect=10, fraction=0.05)
plt.show()
```

```python
plt.contour(Y,f1(X,Y),f1(Y,X), 150, cmap = 'cool')
plt.xlabel('x')                     
plt.ylabel('$\sigma$')
plt.xlim([1,10])
#plt.ylim([2,10])
plt.title('Equipotenciales')
```

### Campo elctrico 
Obtenemos el campo elctrico el cual est dado por:

$-\nabla\phi=\vec{E}=$
 $({\frac{R(x-(R^{2}/s))}{s((x-R^2/s)^2+y^{2})^{3/2}}}+{\frac{s-x}{((s-x)^{2}+y^2)^{3/2}}},{\frac{Ry}{s((x-R^{2}/s)^{2}+y^{2})^{3/2}}}-{\frac{y}{((s-x)^{2}+y^{2})^{3/2}}} ) $

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from pylab import *
from pylab import *

#%config InlineBackend.figure_format = 'retina'

x = np.linspace(1,30,50)
y = np.linspace(1,30,50)
X, Y = np.meshgrid(x, y)
e=8.8e-12
q=-1.6e-19
R=1
s=2*R

def campo_E_V_xy_2(x,y):
    #Ex,Ey = 0,0  (1/(4*np.pi*e))
     #  Ex = R*(x-(R*R/s))/(s*(((x-(R*R/s))**2+y*y)**(3/2)))+(s-x)/(((s-x)**2+y*y)**(3/2))
      #  Ey =(R*y)/(s*(((x-(R*R/s)**2)+y*y)**(3/2)))-y/(((s-x)**2+y**2)**(3/2))
    return (R*(x-(R*R/s))/(s*(((x-(R*R/s))**2+y*y)**(3/2)))+(s-x)/(((s-x)**2+y*y)**(3/2)),(R*y)/(s*(((x-(R*R/s)**2)+y*y)**(3/2)))-y/(((s-x)**2+y**2)**(3/2)))

def graf_campo_E1(x,y,Ex,Ey):
    fig = plt.figure(figsize=(5, 5))
    ax = fig.add_subplot(111)
    ax.streamplot(X, Y, Ex,Ey,density=[1, 1])
    plt.title(r'Campo $\vec{E}$')
    plt.xlabel(r'$\rho$')
    plt.ylabel('z')
    #plt.figure(figsize=(5,20))
    #plt.plot([[0,1], [50, 70]])
  #  plt.xlim([150])
    #plt.ylim([0,d])
   # xlabel(r"$\rho$ [m]",fontsize=18)
    #ylabel('z [m]',fontsize=18)
    ax.set_aspect('equal')
    #show()
Ex1,Ey1 = campo_E_V_xy_2(X,Y)

graf_campo_E1(X, Y, -Ex1,-Ey1)
```
