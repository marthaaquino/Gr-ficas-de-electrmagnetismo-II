$$\vec{E}_{dip}(r,\theta)={\frac{p}{4\pi\epsilon_0r^{3}}}(2\cos\theta\hat{e}_r+\sin\theta\hat{e}_\theta)$$ 

$$\vec{E}=({\frac{2x^{2}}{(x^{2}+y^{2})^{2}}}-\frac{1}{x^{2}+y^{2}},{\frac{2xy}{(x^{2}+y^{2})^{2}}}) $$

$$\vec{E}=-\nabla\phi$$
Graficando el campo el√©ctrico y el potencial tomando:
$\frac{p}{4\pi\epsilon_0}=1$

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from pylab import *
from pylab import *
```

```python
x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)

X,Y = np.meshgrid(x,y)

from matplotlib import cm

def f (x,y):
    
    return (1/((x**2+y**2)**(1/2)))*(x/((x**2+y**2)**(1/2)))


#Axes3D(plt.figure()).plot_surface(X,Y,f(X,Y), rstride=1, cstride=1, cmap = cm.cool)
fig = plt.figure()
ax = Axes3D(fig)
surf=ax.plot_surface(X,Y, f(X,Y), rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=True)
plt.title('Potencial [V]', fontsize=15)
plt.xlabel(r'x [m]')  
plt.ylabel('y [m]')
fig.colorbar(surf, shrink=0.60, aspect=10, fraction=0.05)
plt.show()
```

```python

x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X, Y = np.meshgrid(x, y)


def campo_E_V_xy_1(x,y):
    return (2*x**2/((x**2+y**2)**2))-(1/(x**2+y**2)),2*y*x/((x**2+y**2)**2)

def graf_campo_E(x,y,Ex,Ey):
    fig = plt.figure()#figsize=(10, 10))
    ax = fig.add_subplot(111)
    ax.streamplot(X, Y, Ex,Ey,density=[0.9, 0.9])
    plt.title(r'Campo $\vec{E}$')
    plt.xlabel(r'x')
    plt.ylabel('y')
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    ax.set_aspect('equal')
```

```python
Ex,Ey = campo_E_V_xy_1(X,Y)

graf_campo_E(X, Y, Ex,Ey)
```
