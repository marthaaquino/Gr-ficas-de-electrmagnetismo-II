# Distribución de carga 
## de un dipolo eléctrico junto a una placa metálica

$\sigma(x,y)={\frac{p}{4\pi}}\left(  {\frac{1}{(x^{2}+y^{2})^{5/2}}}(-2x^{2}\cos\alpha-3xy\sin\alpha+y^{2}\sin\alpha) +{\frac{1}{((x-2d)^{2}+y^{2})^{5/2}}}(-2(x-2*d)^{2}+3(x-2d)y\sin\alpha+y^{2}\cos\alpha) \right)$


Sin embargo, para obtener la $\sigma$ en la placa situada en x=d, evaluamos $\sigma(d,y)$ con lo que se obtiene que:

$\sigma(d,y)={\frac{p}{4\pi}}\left(  {\frac{1}{(d^{2}+y^{2})^{5/2}}}(-2d^{2}\cos\alpha-3dy\sin\alpha+y^{2}\sin\alpha) +{\frac{1}{((d-2d)^{2}+y^{2})^{5/2}}}(-2(d-2*d)^{2}+3(d-2d)y\sin\alpha+y^{2}\cos\alpha) \right)$

Esto a partir de una $\alpha$ dada. A continuación se presenta la densidad de carga en la placa.

```python
from matplotlib import pyplot
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
# Función cuadrática.
import math
d=1
x=1
p=1
cte=1/(4*math.pi)
a= 0 #En radianes, 45 grados

y = np.arange(d-10,d+10,0.15)

z=cte*p*((((1/((x*x)+(y*y)))**(5/2))*((-2)*x*x*math.cos(a)+y*y*math.sin(a)))+
                            ((-2)*(x-2*d)*math.cos(a)-(x-2*d)*(x-2*d)*math.sin(a)+2*y*y*math.sin(a))*((1/((x-2*d)**2+y**2)))**(5/2)) 

plt.plot(y,z)
plt.xlabel("y [m]",fontsize=18)
plt.ylabel(r'$\sigma$ [C/m^2]',fontsize=18)
plt.title("Densidad de carga")
plt.show()
#Z =- q/((2*math.pi*((X**2)+z0**2))**(3/2))
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt
import math

from matplotlib import animation, rc
from IPython.display import HTML

d=1
x=1
p=1
cte=1/(4*math.pi)
a=0.785398

fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-1, 1, 0.07)
Y = np.arange(-1, 1, 0.07)
X, Y = np.meshgrid(X, Y)

def fun(x,y):
     return cte*p*((((1/((x*x)+(y*y)))**(5/2))*((-2)*x*x*math.cos(a)+y*y*math.sin(a)))+
                            ((-2)*(x-2*d)*math.cos(a)-(x-2*d)*(x-2*d)*math.sin(a)+2*y*y*math.sin(a))*((1/((x-2*d)**2+y**2)))**(5/2)) 


# Create a figure and a 3D Axes
fig = plt.figure()
ax = Axes3D(fig)

def init():
    # Plot the surface.
    ax.plot_surface(X, Y, fun(X,Y), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
    return fig,

def animate(i):
    # azimuth angle : 0 deg to 360 deg
    ax.view_init(elev=10, azim=i*4)
    return fig,

# Animate
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=100, blit=False)

HTML(anim.to_html5_video())
```
